```python
import pandas as pd

df = pd.read_csv("cotacao.csv", sep=";", decimal=",")
```

Aqui mostraremos a amplitude com a maior e menor cotação no periodo

```python
df['dataHoraCotacao'] = pd.to_datetime(df['dataHoraCotacao'])

df['ano_mes'] = df['dataHoraCotacao'].dt.to_period('M')

colunas = [
    "cotacaoCompra_USD", "cotacaoVenda_USD",
    "cotacaoCompra_EUR", "cotacaoVenda_EUR",
    "paridadeCompra_USDEUR", "paridadeVenda_USDEUR"
]

def calcular_amplitude(grupo):
    return grupo.max() - grupo.min()

amplitude_mensal = df.groupby('ano_mes')[colunas].apply(calcular_amplitude).reset_index()
amplitude_mensal
```

Media movel ajuda a visualizar tendencias de crescimento ou queda.

Aqui analisamos o valor maximo da cotacao de venda e compra do dolar e do euro e tambem a paridade maxima entre o dolar e o euro para o real

```python
df['dataHoraCotacao'] = pd.to_datetime(df['dataHoraCotacao'], errors='coerce')

df['ano_mes'] = df['dataHoraCotacao'].dt.to_period('M')
#Aqui ele pega o ano e o mes com maior cotação
maximaDoDolarPorMes = df.groupby('ano_mes')[['cotacaoCompra_USD', 'cotacaoVenda_USD']].max().reset_index()
maximaDoEuroPorMes = df.groupby('ano_mes')[['cotacaoCompra_EUR', 'cotacaoVenda_EUR', 'paridadeCompra_USDEUR', 'paridadeVenda_USDEUR']].max().reset_index()
```

```python
maximaDoDolarPorMes
```

```python
maximaDoEuroPorMes
```
